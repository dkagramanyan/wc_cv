# Poliamid fracture analysis

Per-group contour metrics are extracted with `PoliamidDataset.generate` and plotted with the `combra.data` plotters. All fitting and plotting lives in the library — this notebook only wires paths together.

```python
from combra import data

poliamid_path = "data/autumn/images"
```

## Generate data

One parquet per per-contour-length cutoff `N` (one row per 250-frame group).

```python
for N in [50, 100, 500, 1000]:
    data.PoliamidDataset(poliamid_path).generate(
        f"poliamid_group250_n{N}.parquet", n_jobs=23, N=N
    )
```

## Fractal dimension distributions

Per-group contour fractal-dimension density with a bimodal-Gaussian fit.

```python
for N in [50, 100, 500, 1000]:
    data.plot_poliamid_fractal(
        f"poliamid_group250_n{N}.parquet",
        step=0.01,
        save=f"poliamid_fractal_n{N}.png",
    )
```

## Contour size distributions

Per-group contour-size density (`len_nodes` / `len_pixels` / `area`) with the candidate distribution fits overlaid.

```python
data.plot_poliamid_contour(
    "poliamid_group250_n1000.parquet",
    metric="area",
    dists=("binomial", "poisson", "gauss", "exponential"),
    step=1,
    save="poliamid_area.png",
)
```
