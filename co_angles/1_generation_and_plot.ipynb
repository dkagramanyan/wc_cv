## Углы

```python
%%time

types_dict = {
    "Ultra_Co11": "средние зерна",
    "Ultra_Co25": "мелкие зерна",
    "Ultra_Co8": "средне-мелкие зерна",
    "Ultra_Co6_2": "крупные зерна",
    "Ultra_Co15": "средне-мелкие зерна",
}

path = './data/generated_images_h5/gen_diff_768x768_N5000.raw.h5'

save_path = './angles/gen_diff_768x768_N5000_msl5'


dataset = data.PobeditDataset(path=path, max_images_num_per_class=5000)

out = dataset.generate_angles(
    save_path=save_path,   
    types_dict=types_dict,
    step=[0.1, 0.5, 1,2,3,4,5],                     
    workers=20,
    angles_tol=3,
    min_segment_len=5.0,
    keep_contours=False,
    chunksize=64
            
)

print("Angles outputs:", out)
```
